### Removing the record with MonthClaimed = 0

```python
# Count how many records
# print(df[df['PolicyNumber'] == 1517])
print(df[df['MonthClaimed'] == '0'])
```

```python
df = df[df['MonthClaimed'] != '0']
```

```python
# Count how many records
# print(df[df['PolicyNumber'] == 1517])
print(df[df['MonthClaimed'] == '0'])
```

```python
df[df['DayOfWeekClaimed']==0]
```

```python
# finding mean of non-zero age
mean_age = int(df[df['Age'] != 0]['Age'].mean())
mean_age
```

```python
# Replace 0 with mean
df.loc[df['Age'] == 0, 'Age'] = mean_age
```

```python
# Verifing any occurences of 0 still
print(f'Mean age used: {mean_age}')
print(df['Age'] == 0)
```

```python
# Saved Cleaned Data 
df.to_csv('../data/processed/cleaned_data.csv', index=False)
```
